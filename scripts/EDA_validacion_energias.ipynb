```python
import pandas as pd
```

```python
df = pd.read_parquet("../data/processed/fact_energia_full_v2.parquet")
```

```python
df.groupby("año")[[
    "gen_solar",
    "gen_eolica",
    "gen_nuclear"
]].sum()
```

```python
df["total_gen"] = (
    df["gen_solar"].fillna(0) +
    df["gen_eolica"].fillna(0) +
    df["gen_nuclear"].fillna(0) +
    df["gen_hidraulica"].fillna(0) +
    df["gen_ciclo_combinado"].fillna(0)
)
```

```python
(df.groupby("año")["gen_eolica"].sum() /
 df.groupby("año")["total_gen"].sum())
```
