```python
import pandas as pd
```

```python
df = pd.read_parquet("../data/raw/esios/precio_mercado/precio_mercado_2020.parquet")
```

```python
# convertir a hora
df["datetime"] = pd.to_datetime(df["datetime"], utc=True)
df["datetime_hora"] = df["datetime"].dt.floor("h")
```

```python
# agrupar
df_hourly = df.groupby("datetime_hora").agg({
    "value": "mean"
}).reset_index()
```

```python
df_hourly.rename(columns={"value": "precio_mercado"}, inplace=True)

df_hourly.head()
```
